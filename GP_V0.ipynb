```python
# Load the Spacy NER model
#nlp = spacy.load("en_core_web_sm")
!python -m spacy download en_core_web_lg

nlp = spacy.load("en_core_web_lg")
```

```python
# Load the dataset
data = pd.read_csv("processed_Tweets_v8.csv")
for tweet in data['Tweet'].head():
    print(tweet)
```

```python
#data['Tweet'] = data['Tweet'].apply(lambda x: re.sub(r'\#\w+', '', x))
for tweet in data['Tweet'].head():
    print(tweet)
```

```python
# Define the sets of entity labels
event_labels = {"EVENT", "GPE", "ORG", "WORK_OF_ART", "PERSON", "NORP", "FAC", "LOC", "PRODUCT", "LAW", "QUANTITY", "MONEY" , "PERCENT"}
time_labels = {"DATE", "TIME", "ORDINAL" }
event_tweets = []
not_event_tweets = []
all_tweets=[]
# Loop over the tweets and perform NER
#for tweet in data['Tweet'].values:
for index, row in data.iterrows():

    doc = nlp(row['Preprocessed Tweet'])

    # Check if there are any entities that belong to both sets of labels
    entities = set(ent.label_ for ent in doc.ents)
    if entities.intersection(event_labels) and entities.intersection(time_labels):
        event_tweets.append(row)
        all_tweets.append(row)
        print("EVENT  :" +str(doc.ents)  + "  " +row['Tweet'] +"\n"+"\n")
        print ("----------------------------------------------------------"+"\n")

    else:
        not_event_tweets.append(row)
        all_tweets.append(row)
        print("NOT EVENT  :" +str(doc.ents) + "  " +row['Tweet'] +"\n"+"\n")
        print ("**********************************************************"+"\n")
```

```python
# Convert the event_tweets list to a dataframe
event_data = pd.DataFrame(event_tweets, columns=["Preprocessed Tweets"])
event_tweets = pd.DataFrame(event_tweets)
event_tweets = event_tweets.dropna()

#stest_not_event_tweets = pd.DataFrame(stest_not_event_tweets)
```

```python
# Prepare the data for classification

x = event_tweets["Preprocessed Tweets"].values
y = event_tweets["Category"].values
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=32)
```

```python
# Vectorize the text data using CountVectorizer
event_tweets = event_tweets.dropna()
vectorizer = CountVectorizer()
vectorizer.fit(x_train)
X_train = vectorizer.transform(x_train)
X_test = vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Train a logistic regression model and make predictions
classifier_lr = LogisticRegression(max_iter=5000)
classifier_lr.fit(X_train, y_train)
y_pred_lr = classifier_lr.predict(X_test)

# Create a DataFrame with the predicted labels
df_predictions_lr = pd.DataFrame({'Category': y_pred_lr})

# Count the occurrences of each category
category_counts_lr = df_predictions_lr['Category'].value_counts()

# Create a pie chart to visualize the category counts
plt.figure(figsize=(8, 6))
category_counts_lr.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Categories (Logistic Regression)')
plt.ylabel('')
plt.show()

# Print the accuracy score
score_lr = classifier_lr.score(X_test, y_test)
print("Logistic Regression Accuracy:", score_lr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Train an SVM model and make predictions
classifier_svm  = SVC()
classifier_svm .fit(X_train, y_train)
y_pred = classifier_svm .predict(X_test)

# Create a DataFrame with the predicted labels
df_predictions = pd.DataFrame({'Category': y_pred})

# Count the occurrences of each category
category_counts = df_predictions['Category'].value_counts()

# Create a pie chart to visualize the category counts
plt.figure(figsize=(8, 6))
category_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Categories (SVM)')
plt.ylabel('')
plt.show()

# Print the accuracy score
score_svm  = classifier_svm.score(X_test, y_test)
print("SVM Accuracy:", score_svm)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest model and make predictions
classifier_rf = RandomForestClassifier()
classifier_rf.fit(X_train, y_train)
y_pred_rf = classifier_rf.predict(X_test)

# Create a DataFrame with the predicted labels
df_predictions_rf = pd.DataFrame({'Category': y_pred_rf})

# Count the occurrences of each category
category_counts_rf = df_predictions_rf['Category'].value_counts()

# Create a pie chart to visualize the category counts
plt.figure(figsize=(8, 6))
category_counts_rf.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Categories (Random Forest)')
plt.ylabel('')
plt.show()

# Print the accuracy score
score_rf = classifier_rf.score(X_test, y_test)
print("Random Forest Accuracy:", score_rf)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Train a Naive Bayes model and make predictions
classifier_nb = MultinomialNB()
classifier_nb.fit(X_train, y_train)
y_pred_nb = classifier_nb.predict(X_test)

# Create a DataFrame with the predicted labels
df_predictions_nb = pd.DataFrame({'Category': y_pred_nb})

# Count the occurrences of each category
category_counts_nb = df_predictions_nb['Category'].value_counts()

# Create a pie chart to visualize the category counts
plt.figure(figsize=(8, 6))
category_counts_nb.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Categories (Naive Bayes)')
plt.ylabel('')
plt.show()

# Print the accuracy score
score_nb = classifier_nb.score(X_test, y_test)
print("Naive Bayes Accuracy:", score_nb)
```

```python
# Load the new dataset with only one column (tweet)
new_data = pd.read_csv("ready.csv"  )


# Initialize the list to store the results
new_results = []

# Iterate over the tweets in the new dataset and perform event detection and classification
id_counter = 1

# Iterate over the tweets in the new dataset and perform event detection and classification
for index, row in new_data.iterrows():
    tweet = row['Preprocessed Tweets']
    Tweets = row['Tweet']
    User = row['User']
    Date = row['Date']
    Time = row['Time']

    # Perform event detection
    doc = nlp(tweet)
    entities = set(ent.label_ for ent in doc.ents)
    if entities.intersection(event_labels) and entities.intersection(time_labels):
        # Event detected, perform classification
        tweet_vector = vectorizer.transform([tweet])
        category = classifier_svm.predict(tweet_vector)[0]

        # Generate a unique ID for the tweet
        tweet_id = id_counter
        id_counter += 1

        # Store the result (tweet, category)
        result = {
            'id': tweet_id,
            'tweet': Tweets,
            'category': category,
            'User': User,
            'likes': 0,
            'Date' : Date,
            'Time' : Time,
        }
        new_results.append(result)

# Print the new results
for result in new_results:
    print("Tweet ID:", result['id'])
    print("Tweet:", result['tweet'])
    print("Category:", result['category'])
    print("Likes:", result['likes'])
    print("User:", result['User'])
    print("Date:", result['Date'])
    print("Time:", result['Time'])
    print("-----------------------------")
```
